```python
import ollama
def generate_insights(titanic_summary):
    prompt=f"Analyze the dataset and priovide insights:\n\n{titanic_summary}"
    response=ollama.chat(model="mistral",messages=[{"role":"user","content":prompt}])
    return response['message']['content']

summary=titanic.describe().to_string()
insights = generate_insights(summary)
print("\n AI-Generated Insights:\n",insights)
```

```python
import gradio as gr
def eda_analysis(file):
    df=pd.read_csv(file.name)
    summary=df.describe().to_string()
    insights=generate_insights(summary)
    return insights

demo=gr.Interface(fn=eda_analysis,inputs='file',outputs="text",title="AI-Powered EDA with Mistral")


demo.launch(share=True)
```
